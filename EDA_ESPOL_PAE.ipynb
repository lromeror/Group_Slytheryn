```python
#Solo el tiempo 
def get_time(df,name_column):
  i=0
  l_dura = []
  for dura in df[name_column]:
    dura = str(dura).split()[2]
    l_dura.append(dura)
  df[name_column] = l_dura
  return df[name_column]
df_H_p['DURACION'] = get_time(df_H_p,'DURACION')
```

```python
#Pasar a int las columnas
def cambiarPunt(df,fea):
    w=0
    for i in df[fea]:
        rem = int(i.split(',')[0])
        df.loc[w,fea] = rem
        w+=1
    return df[fea]
for name_col in ['PARALELO','CAPACIDAD','CUPOPLANIFICADO']:
  df_H_p[name_col] = cambiarPunt(df_H_p,name_col)
```
